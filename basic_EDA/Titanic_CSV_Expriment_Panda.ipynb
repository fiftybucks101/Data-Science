```python
age_group = ['children','adults','elderly']

no_of_children = df[(df["Age"] >=0) & (df["Age"] <= 19)]["PassengerId"].count()

no_of_adult = df[(df["Age"] >=20) & (df["Age"] <= 39)]["PassengerId"].count()

no_of_elderly = df[(df["Age"] >=40)]["PassengerId"].count()

alive_children = df[(df["Age"] >=0) & (df["Age"] <= 19) & (df["Survived"] == 1)]["PassengerId"].count()

alive_adult = df[(df["Age"] >=20) & (df["Age"] <= 39) & (df["Survived"] == 1)]["PassengerId"].count()

alive_elderly = df[(df["Age"] >=40) & (df["Survived"] == 1)]["PassengerId"].count()

# survivour_rate = {
#     'Children': alive_children/no_of_children*100,
#     'Adults':alive_adult/no_of_adult*100,
#     'Elderly':alive_elderly/no_of_elderly*100
# }

survivour_rate = {
    "Age Range":["1-19","20-39","Above 40"],
    "Total":[no_of_children,no_of_adult,no_of_elderly],
    "Survivour %": [alive_children/no_of_children*100,alive_adult/no_of_adult*100,alive_elderly/no_of_elderly*100],
    }

age_group_survivour = pd.DataFrame(survivour_rate,index=age_group)

age_group_survivour


# age_group_survivour

```

```python
age_group_survivour.plot(kind = "barh")

plt.title("Titanic Age Range Survivour %")

plt.show()
```

```python
bins = [0,25,70,150,600] # bins used for catagorizing such as [0-100],[100-200],[300-600]
labels = ["3rd Class",'2nd Class','1st Class','VIP']

df['Class'] = pd.cut(df["Fare"],bins=bins,labels=labels,right=False)

fare_range=["0-24","25-69","69-149","149-Max"]



survivour_by_class = df.groupby('Class')['Survived'].mean()*100

total_passenger_by_class = df.groupby('Class')['Survived'].count()

survivour_by_class_df=pd.DataFrame({"Total passenger":total_passenger_by_class,"Fare Range":dict(zip(labels,fare_range)),"Survival Rate":survivour_by_class})

reversed_index = survivour_by_class_df.index.tolist()[::-1]

survivour_by_class_df = survivour_by_class_df.reindex(reversed_index)

survivour_by_class_df

```

```python
# Example data
ages = [10, 20, 25, 35, 45, 50, 60, 70, 80]

# Define bins and labels
bins = [0, 20, 40, 60, 100]
labels = ['0-20', '20-40', '40-60', '60-100']

# Use pd.cut with right=True
age_groups_right_true = pd.cut(ages, bins=bins, labels=labels, right=True)
print("Bins with right=True:")
print(age_groups_right_true)

# Use pd.cut with right=False
age_groups_right_false = pd.cut(ages, bins=bins, labels=labels, right=False)
print("\nBins with right=False:")
print(age_groups_right_false)
```
