```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np

import yfinance as yahooFinance

meta_data = yahooFinance.Ticker("META")

meta_history = meta_data.history(period="1mo")
```

```python
meta_df = pd.DataFrame(data=meta_history)
meta_df.head()
```

```python
meta_df.info()
```

```python
meta_df.describe()
```

```python
meta_df.drop(columns=['Dividends', 'Stock Splits'], inplace=True)
```

```python
meta_df.head()
```

```python
meta_df.isna()
```

```python
meta_df.isnull()
```

```python
meta_df.info()
```

```python
meta_df.columns = [col.lower() for col in meta_df.columns]
```

```python
meta_df.index.name = meta_df.index.name.lower()
```

```python
meta_df.head()
```

```python
meta_df.head()
```

```python
df_length = len(meta_df)
candle_color = []

for i in range(0,df_length,1):
  if meta_df['close'].iloc[i] > meta_df['open'].iloc[i]:
    candle_color.append('green')
  else:
    candle_color.append('red')
```

```python
len(candle_color)
```

```python
meta_df['candle_color'] = candle_color

meta_df.head()
```

```python
sns.relplot(data=meta_df, x="date", y = "close", aspect = 3, size = "volume", palette=candle_color, hue='candle_color')
```

```python
meta_df['volume'].describe()
```

```python
sns.boxplot(data=meta_df, x='candle_color',y='volume')
```

```python
meta_df.groupby('candle_color').count()

meta_df.reset_index(inplace=True)
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(data = meta_df, x='date', y = 'volume', hue='candle_color',width=0.5)

plt.xticks(rotation=45)
```
